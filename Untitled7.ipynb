```python
vers_rgx = re.compile("v[0-9]*\.[0-9]*\.[0-9]*")
ffr_rgx_1 = re.compile("\-[0-9]*\,[0-9]*")
ffr_rgx_2 = re.compile("\-[0-9]* ")
tfr_rgx_1 = re.compile("\+[0-9]*\,[0-9]*")
tfr_rgx_2 = re.compile("\+[0-9]* ")


# https://en.wikipedia.org/wiki/Diff#Unified_format
# https://stackoverflow.com/questions/2529441/how-to-read-the-output-from-git-diff
def diff_parser(diff):
    """
    diff must be in a "xx xx @@ yyy" format
    """
    if (diff == "") or (diff is None):
        return None
    diff_loc = diff.split("@@")[0]
    try:
        ffr_startline, ffr_endline = ffr_rgx_1.search(diff_loc).group(0).split(",")
    except:
        ffr_startline = ffr_rgx_2.search(diff_loc).group(0)
        ffr_endline = 1
    ffr_startline = ffr_startline[1:]
    try:
        tfr_startline, tfr_endline = tfr_rgx_1.search(diff_loc).group(0).split(",")
    except:
        tfr_startline = tfr_rgx_2.search(diff_loc).group(0)
        tfr_endline = tfr_startline
    tfr_startline = tfr_startline[1:]
    added_loc = []
    deleted_loc = []
    from_file = int(ffr_startline)
    to_file = int(tfr_startline)
    for x in diff.split("\n")[1:]:
        if len(x) == 0:
            from_file += 1
            to_file += 1
            continue
        if x[0] == "+":
            added_loc.append(to_file)
            to_file += 1
            continue
        if x[0] == "-":
            deleted_loc.append(from_file)
            from_file += 1
            continue
        from_file += 1
        to_file += 1
    return added_loc, deleted_loc, "\n".join(diff.split("\n")[1:])


# https://en.wikipedia.org/wiki/Diff#Unified_format
# https://stackoverflow.com/questions/2529441/how-to-read-the-output-from-git-diff
def diff_file_parser(diff_file):
    """
    diff must be unified format
    """
    parsed = []
    for diff in re.split(r"^@@|\n@@", diff_file)[1:]:
        added_loc, deleted_loc, diff = diff_parser(diff)
        parsed.append({
            'added_loc': added_loc,
            'deleted_loc': deleted_loc,
            'patch': diff})
    return parsed


def ext_parser(fn):
    try:
        file = fn.split("/")[-1]
        elts = file.split(".")
        if len(elts) == 1:
            return ''
        else:
            return elts[-1]
    except:
        return None


def vers_parser(msg):
    try:
        return vers_rgx.search(msg).group(0)
    except:
        return None


def language_detection(ext):
    return ext2lg.get(ext, 'UNK')


def github_jwt_refresher(devlens_token, crl=None):
    if crl:
        reset_timestamp = calendar.timegm(crl.reset.timetuple())
        sleep_time = reset_timestamp - calendar.timegm(
            time.gmtime()) + 5  # add 5 seconds to be sure the rate limit has been reset
        logging.info(f"Reset is in {sleep_time:.3g} seconds.")
        if sleep_time > 0.0:
            logging.info(f"Waiting for {sleep_time} seconds...")
            time.sleep(sleep_time)
            logging.info("Done waiting - resume!")
    github_token = requests.post(settings.BACKEND_HOST + '/user_configuration/fetch_github_token/',
                                 data=json.dumps(
                                     {'devlens_token': devlens_token
                                      }),
                                 headers={"Accept": "application/json",
                                          "Content-Type": "application/json"}).json()['github_token']
    g = Github(github_token, per_page=100, retry=10)
    return g
```

```python
def batch_indexing_task(self, index_name, devlens_token, batch_shas, commit2branches, repository=None):

# Connect to Elasticsearch
es = Elasticsearch("https://" + settings.ELASTICSEARCH_HOST,
                   http_auth=(settings.ELASTICSEARCH_USERNAME,
                              settings.ELASTICSEARCH_PASSWORD), timeout=240)

# Let's Populate the diff index
def diff_iterator(index_name, devlens_token, batch_shas, commit2branches, repository=None):

    github_token = requests.post(settings.BACKEND_HOST + '/user_configuration/fetch_github_token/',
                                 data=json.dumps(
                                     {'devlens_token': devlens_token
                                      }),
                                 headers={"Accept": "application/json",
                                          "Content-Type": "application/json"}).json()['github_token']
    g = Github(github_token, per_page=100, retry=10)

    diff_count = 0
    commit_count = 0
    diff_t0 = time.time()
    commit_t0 = time.time()
    for r_name, r_id, sha in batch_shas:
        if repository is None or repository == r_name:
            try:
                commit = g.get_repo(r_id).get_commit(sha)
            except RateLimitExceededException:
                core_rate_limit = g.get_rate_limit().core
                g = github_jwt_refresher(devlens_token, crl=core_rate_limit)
                diff_t0 = time.time()
                commit_t0 = time.time()
                commit = g.get_repo(r_id).get_commit(sha)
            except BadCredentialsException:
                g = github_jwt_refresher(devlens_token)
                diff_t0 = time.time()
                commit_t0 = time.time()
                commit = g.get_repo(r_id).get_commit(sha)
            try:
                commit_count += 1
                if commit_count % 1000 == 0:
                    commit_t1 = time.time()
                    logging.info("1000 Commits indexed in {0} s".format(commit_t1 - commit_t0))
                    commit_t0 = time.time()
                    #if settings.ENV == 'development':
                    #    break
                # TODO : rajouter branch?
                branches = commit2branches[commit.sha]
                # TODO : rajouter namespace?
                timestamp = commit.commit.author.date
                parents = commit.parents
                author = commit.author
                author_avatar_url = author.avatar_url if author else ""
                message = commit.commit.message
                commit_id = commit.sha
                repository = r_name
                versioning = vers_parser(message)
                url = commit.html_url
                try:
                    for changed_file in commit.raw_data["files"]:
                        filename = changed_file["filename"]
                        extension = ext_parser(filename)
                        language = language_detection(extension)
                        """
                        diff must be in a "xx xx @@ yyy" format in the diff parser
                        """
                        try:
                            if 'patch' in changed_file:
                                for diff in diff_file_parser(changed_file['patch']):
                                    added_loc = diff['added_loc']
                                    deleted_loc = diff['deleted_loc']
                                    plus_volume = len(added_loc)
                                    minus_volume = len(deleted_loc)
                                    plus_line_numbers = added_loc
                                    minus_line_numbers = deleted_loc
                                    code = diff['patch']
                                    filename = filename.lstrip().rstrip()
                                    diff_doc = {
                                        "company": author.company if author else "",
                                        "repository": repository,
                                        "@timestamp": timestamp,
                                        "commit_id": commit_id,
                                        "parent_id": [p.sha for p in parents],
                                        "author": author.login if author else commit.commit.author.name,
                                        "message": message,
                                        "filename": filename,
                                        "extension": extension,
                                        "plus_volume": plus_volume,
                                        "plus_line_numbers": plus_line_numbers,
                                        "minus_volume": minus_volume,
                                        "minus_line_numbers": minus_line_numbers,
                                        "code": code,
                                        "versioning": versioning,
                                        "day": timestamp.day,
                                        "hour": timestamp.hour,
                                        "year": timestamp.year,
                                        "month": timestamp.month,
                                        'branches': [repository + '/' + br for br in branches],
                                        'risk_weight': 1,
                                        'language': language,
                                        "url": url,
                                        "author_avatar_url": author_avatar_url
                                    }
                            else:
                                diff_doc = {
                                    "company": author.company if author else "",
                                    "repository": repository,
                                    "@timestamp": timestamp,
                                    "commit_id": commit_id,
                                    "parent_id": [p.sha for p in parents],
                                        "author": author.login if author else commit.commit.author.name,
                                    "message": message,
                                    "filename": filename,
                                    "extension": extension,
                                    "versioning": versioning,
                                    "day": timestamp.day,
                                    "hour": timestamp.hour,
                                    "year": timestamp.year,
                                    "month": timestamp.month,
                                    'branches': [repository + '/' + br for br in branches],
                                    'risk_weight': 1,
                                    'language': language,
                                        "url": url,
                                        "author_avatar_url": author_avatar_url
                                }
                            diff_count += 1
                            if diff_count % 1000 == 0:
                                diff_t1 = time.time()
                                logging.info("1000 Diffs indexed in {0} s".format(diff_t1 - diff_t0))
                                diff_t0 = time.time()
                            yield {
                                "_index": index_name,
                                "_type": "diff",
                                "_source": diff_doc
                            }
                        except RateLimitExceededException:
                            core_rate_limit = g.get_rate_limit().core
                            g = github_jwt_refresher(devlens_token, crl=core_rate_limit)
                        except BadCredentialsException:
                            g = github_jwt_refresher(devlens_token, crl=core_rate_limit)
                        except Exception as e:
                            logging.info(type(e), e)
                except RateLimitExceededException:
                    core_rate_limit = g.get_rate_limit().core
                    g = github_jwt_refresher(devlens_token, crl=core_rate_limit)
                except BadCredentialsException:
                    g = github_jwt_refresher(devlens_token, crl=core_rate_limit)
                except Exception as e:
                    logging.info(type(e), e)
            except RateLimitExceededException:
                core_rate_limit = g.get_rate_limit().core
                g = github_jwt_refresher(devlens_token, crl=core_rate_limit)
            except BadCredentialsException:
                g = github_jwt_refresher(devlens_token, crl=core_rate_limit)
            except Exception as e:
                logging.info(type(e), e)

logging.info('\n##########################'
             '\nINDEXING\n'
             '##########################')

indexing_t0 = time.time()

success, info = bulk(client=es,
                     actions=diff_iterator(index_name, devlens_token, batch_shas, commit2branches, repository),
                     chunk_size=500,
                     raise_on_error=True)

indexing_t1 = time.time()


logging.info('\n##########################'
             f'\nINDEXING DONE IN {indexing_t1- indexing_t0}\n'
             '##########################')

return Response()
```

```python
def updating_task(self, index_name):
logging.info('\n##########################'
             '\nUPDATING\n'
             '##########################')

# Update : Computing NW/C/HO/LR
match_all_sorted = {
    "sort":
        {"@timestamp": {"order": "asc"}}
    ,
    "query": {
        "match_all": {}
    }
}

def simple_last_modifying_commit(diff, ts, plus_line_number):
    query = {
        "sort": {
            "@timestamp": {
                "order": "desc"
            }
        },
        "query": {
            "bool": {
                "filter": [
                    {
                        "range": {
                            "@timestamp": {
                            }
                        }
                    },
                    {
                        "bool": {
                            "must": [
                                {
                                    "match": {
                                    }
                                }
                            ]
                        }
                    },
                    {
                        "bool": {
                            "must": [
                                {
                                    "term": {

                                    }
                                }
                            ]
                        }
                    }
                ]
            }
        },
        "size": "1"
    }
    query["query"]["bool"]["filter"][1]["bool"]["must"][0]["match"]["filename"] = diff["_source"]["filename"]
    query["query"]["bool"]["filter"][2]["bool"]["must"][0]["term"]["plus_line_numbers"] = plus_line_number
    query["query"]["bool"]["filter"][0]["range"]["@timestamp"]["lt"] = ts
    return query

def update_iterator(index_name):
    diff_count = 0
    diff_t0 = time.time()
    for diff in scan(es,
                     query=match_all_sorted,
                     index=index_name,
                     doc_type="diff",
                     preserve_order=True,
                     scroll='50m'
                     ):
        N = 0  # New
        I = 0  # Iteration
        H = 0  # Help
        DC = 0  # Debt Cleaning
        current_t = datetime.fromisoformat(diff["_source"]["@timestamp"])
        author_t = diff["_source"]["author"]
        try:
            for mln in diff["_source"]["minus_line_numbers"]:
                res = es.search(index=index_name, body=simple_last_modifying_commit(diff=diff,
                                                                                    ts=current_t,
                                                                                    plus_line_number=mln),
                                doc_type="diff")
                try:
                    t = datetime.fromisoformat(res["hits"]["hits"][0]["_source"]["@timestamp"])
                except Exception as e:
                    #logging.info(e)
                    #logging.info(res)
                    continue
                author = res["hits"]["hits"][0]["_source"]["author"]
                if (current_t - t).days < 30:
                    if author != author_t:
                        H += 1
                    else:
                        I += 1
                else:
                    DC += 1
            N += diff["_source"]["plus_volume"]
        except Exception as e:
            #logging.info(e)
            continue
        devprint = N + DEVPRINT_ITERATION * I + DEVPRINT_HELP * H + DEVPRINT_DEBT_CLEANING * DC
        diff_doc = {
            "new": N,
            "debt_cleaning": DC,
            "iteration": I,
            "help": H,
            "devprint": devprint,
            "risk": diff["_source"]["risk_weight"] * devprint}
        diff_count += 1
        if diff_count % 1000 == 0:
            diff_t1 = time.time()
            logging.info("1000 Diffs indexed in {1} s".format(diff_count, diff_t1 - diff_t0))
            diff_t0 = time.time()
        yield {
            "_index": index_name,
            "_type": "diff",
            "doc": diff_doc,
            "_id": diff["_id"],
            "_op_type": "update"
        }

time.sleep(1)

updating_t0 = time.time()

success, info = bulk(client=es,
                     actions=update_iterator(index_name),
                     chunk_size=2500,
                     raise_on_error=True,
                     )

updating_t1 = time.time()

logging.info('\n##########################'
             f'\nUPDATING DONE IN {updating_t1 - updating_t0}\n'
             '##########################')
```
