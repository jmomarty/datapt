```python
import pandas as pd

df = pd.io.parsers.read_csv(
    'https://raw.githubusercontent.com/rasbt/pattern_classification/master/data/wine_data.csv',
    header=None,
    )
```

```python
df
```

```python
df.describe().T
```

```python
X = df.values[:, 1:]
y = df.values[:, 0]
```

```python
y.shape
```

```python
from sklearn import preprocessing
std_scale = preprocessing.StandardScaler().fit(X)
```

```python
X_std = std_scale.transform(X)
```

```python
pd.DataFrame(X_std).describe().T
```

```python
import numpy as np
```

```python
np.unique(y)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2).fit(X)
X_pca = pca.transform(X)
```

```python
X_pca.shape
```

```python
pca_std = PCA(n_components=2).fit(X_std)
X_pca_std = pca_std.transform(X_std)
```

```python
from matplotlib import pyplot as plt

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10,4))


for l,c,m in zip(range(1,4), ('blue', 'red', 'green'), ('^', 's', 'o')):
    ax1.scatter(X_pca[y==l, 0], X_pca[y==l, 1],
        color=c,
        label='class %s' %l,
        alpha=0.5,
        marker=m
        )

for l,c,m in zip(range(1,4), ('blue', 'red', 'green'), ('^', 's', 'o')):
    ax2.scatter(X_pca_std[y==l, 0], X_pca_std[y==l, 1],
        color=c,
        label='class %s' %l,
        alpha=0.5,
        marker=m
        )

ax1.set_title('Transformed NON-standardized training dataset after PCA')    
ax2.set_title('Transformed standardized training dataset after PCA')    

for ax in (ax1, ax2):
    ax.set_xlabel('1st principal component')
    ax.set_ylabel('2nd principal component')
    ax.legend(loc='upper right')
    ax.grid()
plt.tight_layout()

plt.show()  
```

```python
pca_std.components_.shape
```

```python
pca_std.explained_variance_ratio_
```
